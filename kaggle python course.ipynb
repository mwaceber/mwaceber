```python
def inspect(x) :
    if x == 0:
        print(x, "is zero")
    elif x > 0:
        print(x, "is positive")
    elif x < 0:
        print(x, "is negative")
    else:
        print(x, "is unlike anything I've ever seen...")

inspect(0)
inspect(-15)
```

```python
def mult_by_five(x) :
    return 5 * x

def call(fn, arg) :
    """Call fn on arg"""
    return fn(arg)

def squared_call(fn, arg) :
    """Call fn on result of calling fn on arg"""
    return fn(fn(arg))

print(
    call(mult_by_five, 1), 
    squared_call(mult_by_five, 1), sep= '\n'
)
```

```python
def least_difference(a, b, c) :
    """Return the smallest difference between any two numbers among
    a, b and c
    
    >>> least_difference(1, 5, -5)
    4
    """
    diff1 = abs(a - b)
    diff2 = abs(b - c)
    diff3 = abs(a - c)
    return min(diff1, diff2, diff3)
print(
    least_difference(1, 10, 100), 
    least_difference(1, 10, 10), 
    least_difference(5, 6, 7)
)
```
